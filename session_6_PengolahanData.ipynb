```python
def fill_missing_by_pclas(dataset, column):
  for i in range(1,4):
    median = dataset[dataset['Pclass'] == i][column].median()
    dataset[column] = dataset[column].fillna(median)
    return dataset
```

```python
fill_missing_by_pclas(titanic, 'Age')
```
