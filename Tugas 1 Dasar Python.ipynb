##Missing value pada Age
Menggunakan median karena data tidak berdistribusi normal dan terdapat skewness

```python
median_age=df_titanic_train['age'].median()
print('Median Age',median_age)
df_titanic_train['age']=df_titanic_train['age'].fillna(median_age)
```

```python
df_titanic_train.isnull().sum()
```

```python
df_titanic_train.hist('fare')
```

##Missing Value pada Fare
Menggunakan median karena data tidak berdistribusi normal dan terdapat skewness

```python
median_fare=df_titanic_train['fare'].median()
print('Median fare',median_fare)
df_titanic_train['fare']=df_titanic_train['fare'].fillna(median_fare)
```

```python
df_titanic_train.isnull().sum()
```

## Missing Value pada Embarked
pengisiaan ini menggunakan modus karena data bersifat kategori

```python
Modus_embarked=df_titanic_train['embarked'].mode()[0]
print('Modus Embarked', Modus_embarked)
df_titanic_train['embarked']=df_titanic_train['embarked'].fillna(Modus_embarked)
```

```python
df_titanic_train.isnull().sum()
```

```python
data1=pd.read_csv('/content/drive/MyDrive/PDA_DataSource/Pengolahan_Data/telecom_revenue.csv')
data2=pd.read_csv('/content/drive/MyDrive/PDA_DataSource/Pengolahan_Data/telecom_usage.csv')
join_data= data1.join(data2.set_index('CustomerID'),on='CustomerID')
join_data
```

```python
df1=join_data[(join_data['MonthlyRevenue']> 10 ) & (join_data['Occupation'].isin(['Professional','Student','Crafts']))]
df1
```

```python
df1.reset_index(drop=True)
```

```python
Monthlyrevenue=df1[df1['Churn']=='No']
ratarevenue=Monthlyrevenue['MonthlyRevenue'].mean()
ratarevenue
```

```python
df1.groupby(['Churn']).agg({'MonthlyRevenue':'mean'})
```

```python
df2=join_data[(join_data['CustomerCareCalls']>10)&(join_data['MonthlyRevenue']<=ratarevenue)]
df2.reset_index(drop=True)
```

```python
df3=join_data.iloc[[2331,3165,3867,4030,4570],0:].reset_index(drop=True)
df3
```

```python
union=pd.concat([df1,df2,df3])
union
```

```python
union_g=union.groupby(['Occupation','Churn']).agg({'CustomerID':'count'}).reset_index()
union_g['Totalcust']=union_g.groupby(['Occupation'])['CustomerID'].transform('sum')
union_g['Churn rate']=union_g['CustomerID']/union_g['Totalcust']
union_g=union_g[(union_g['Churn']=='Yes')& (union_g['Churn rate']>=0.4)]
union_g
```

Jadi Occupation dengan Churn rate >= 0.4 adalah Clerical dengan Churn rate 1.0
