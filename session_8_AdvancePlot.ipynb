```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/drive/MyDrive/PDA_DataSource/Data Visualization/percent-bachelors-degrees-women-usa.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Jawaban Eisya Adila
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='lower right')
plt.show()
```

```python
# tips
# definisikan kolom2 yang akan digunakan
year = df['Year']
physical_science = df['Physical Sciences']
computer_science = df['Computer Science']
psychology = df['Psychology']
business = df['Business']
```

```python
plt.plot(year, computer_science) # == plt.plot(df['Year], df['Computer Science'])
```

## Menggunakan subplot

```python
plt.figure(figsize=(10, 6))
plt.plot(year, business)
plt.show()
```

```python
# satu visual
plt.subplots(1,1, figsize=(5,3))
plt.plot(year, business)
plt.show()
```

```python
# variabel subplot
fig, ax = plt.subplots(1,1, figsize=(5,3))
ax.plot(year, business)
sns.lineplot(x=year, y=physical_science, ax=ax)
```

```python
# multiple visual
fig, ax = (plt.subplots(2,1, figsize=(8,5)))
ax[0].plot(year, business)
sns.lineplot(x=year, y=physical_science, ax=ax[1], color='red')
plt.show()
```

## Label

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, psychology)
sns.lineplot(x=year, y=computer_science, ax=ax[1], color='red')
ax[0].set_title('psychology over year', fontsize=16)
ax[1].set_title('computer science over year', fontsize=16)
fig.tight_layout(pad=5) # memberikan space antara visualisasi

# memberikan label
ax[0].set_xlabel('Year', fontsize=10)
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, psychology)
ax[1].plot(year, computer_science, color='red')
ax[0].set_title('Psychology over year', fontsize=16)
ax[1].set_title('Computer science over year', fontsize=16)
fig.tight_layout(pad=5) # memberikan space antara visualisasi

# memberikan label
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas
plt.show()
```

## X-Y Limit

```python
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='lower right')
plt.xlim(1985,1995)
plt.ylim(25, 40)
plt.xticks(list(range(1985,1996,5)))
plt.yticks(list(range(25,41,5)))
plt.show()
```

## Legenda

```python
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='center')
plt.xlim(1985,1995)
plt.ylim(25, 40)
plt.xticks(list(range(1985,1996,5)))
plt.yticks(list(range(25,41,5)))
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, psychology, label='psy')
ax[0].plot(year, computer_science, color='red', label='cs')
ax[0].set_title('Psychology over year', fontsize=16)
ax[0].legend(loc='lower right')

ax[1].set_title('Computer science over year', fontsize=16)
fig.tight_layout(pad=5) # memberikan space antara visualisasi

# memberikan label
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas

# memberikan legenda
ax[0].legend(loc='lower right')
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6), sharey=True, sharex=True)
ax[0].plot(year, psychology, label='psy')
ax[1].plot(year, computer_science, color='red', label='cs')
ax[0].set_title('Psychology over year', fontsize=16)
ax[0].legend(loc='lower right')

ax[1].set_title('Computer science over year', fontsize=16)
fig.tight_layout(pad=5) # memberikan space antara visualisasi

# memberikan label
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas

# memberikan legenda
ax[0].legend(loc='lower right')
plt.show()
```

tambahkan teks di ax[0] dan juga ax[1]. text nya bebas

```python
fig, ax = plt.subplots(2,1, figsize=(10,6), sharey=True, sharex=True)
ax[0].plot(year, psychology, label='psy')
ax[1].plot(year, computer_science, color='red', label='cs')
ax[0].set_title('Psychology over year', fontsize=16)
ax[0].legend(loc='lower right')

ax[1].set_title('Computer science over year', fontsize=16)
fig.tight_layout(pad=5) # memberikan space antara visualisasi

# memberikan label
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas

# memberikan legenda
ax[0].legend(loc='lower right')

# memberikan text
ax[0].text(x=1980, y=50, s='ini text', ha='center')
ax[1].text(x=1980, y=50, s='ini text', ha='center')
plt.show()
```
